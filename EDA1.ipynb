```python
import numpy as np                # 넘파이 패키지 임포트
import pandas as pd               # 판다스 패키지 임포트
import matplotlib.pyplot as plt   # 맷플롯립 패키지 임포트

# 깃허브에 있는 csv 파일을 읽어와서 데이터프레임 df로 넣는다.
df = pd.read_csv("https://raw.githubusercontent.com/YoungjinBD/dataset/main/titanic.csv")
```

```python
df.info()
```

```python
# 변수 타입 변환
df["Survived"] = df["Survived"].astype(str)
df["Pclass"] = df["Pclass"].astype(str)
```

```python
df.info() 
```

```python
df.describe(include='all')
```

```python
# groupby(), size()를 이용하여 Pclass 빈도수 구하기
grouped = df.groupby("Pclass")
grouped.size()
```

```python
# Histogram
plt.hist(df["Fare"])
plt.show()

# 데이터 분리
data_0 = df[df["Survived"]=="0"]["Fare"] # 사망자의 요금 데이터
data_1 = df[df["Survived"]=="1"]["Fare"] # 생존자의 요금 데이터

# Boxplot – Survived 값에 따른 Fare 값의 분포를 보여준다.
fig, ax = plt.subplots()
ax.boxplot([data_0, data_1])
plt.show()
```

```python
# 빈도 수 계산 (성별)
grouped = df.groupby("Sex")
print(grouped.size())

# 성별 생존자 수
data_0 = df[df["Sex"]=="female"]["Survived"]          # 여성의 생존 데이터
grouped = pd.DataFrame(data_0).groupby("Survived")
print(grouped.size())

data_1 = df[df["Sex"]=="male"]["Survived"]            # 남성의 생존 데이터
grouped = pd.DataFrame(data_1).groupby("Survived")
print(grouped.size())
```
